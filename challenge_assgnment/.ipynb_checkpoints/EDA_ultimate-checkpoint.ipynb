```python
import json
import pandas as pd
from datetime import datetime 
from datetime import timedelta
import numpy as np
from collections import defaultdict
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
f_1 = open('logins.json')
```

```python
data_1 = json.load(f_1)
```

```python
df_login = pd.DataFrame(data_1)
```

```python
df_login.dtypes
```

```python
df_login.index = pd.to_datetime(df_login['login_time'])
```

```python
df_login.resample(rule='15T').count()
```

```python
plt.figure(figsize=(12,6))
df_login.resample(rule='15T').count().plot();
```

```python
df_agg = df_login.resample(rule='15T').count()
```

```python
#now each row contains 15 min data, we can plot for 96 window to see per day pattern
```

```python
df_agg.rolling(96).mean().plot()
```

```python
#lets plot only a month day wise
```

```python
df_agg.loc['1970-02'].rolling(96).mean().plot(figsize=(12,6))
```

```python
#plotting the trend over a week 
```

```python
datetime(1970,3,1).weekday()
```

```python
unsmoothed = df_agg.loc['1970-03-01':'1970-03-07']['login_time']
trend = unsmoothed.rolling(48).mean()
pd.DataFrame({'unsmoothed':unsmoothed,'smoothed':trend}).plot(figsize=(12,6))
```

```python
#from above graph its clear that demand is higher at inetrval (7-9,14-16,21-23)
#lets check for wat is the day on these days to check demand on weekend and weekdays
```
